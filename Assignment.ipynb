```python
import csv
import json
from json import JSONEncoder
```

```python
class setEncoder(JSONEncoder):
        def default(self, obj):
            return list(obj)
```

```python
with open ('C:\\Users\\admin\\Desktop\\Assignment\\StudentsPerformance.csv','r') as f:
    reader = csv.reader(f)
    next(reader) #skips header
    
    data = []
    
    for row in reader:
            data.append({"gender":row[0],"race/ethnicity":row[1],"parental level of education":row[2],"lunch":row[3],"test preparation course":row[4],"math score":row[5],"reading score":row[6],"writing score":row[7]})
    
    print(data)
```

```python
with open ("C:\\Users\\admin\\Desktop\\Assignment\\StudentsPerformance.json","w") as f:
    json_data = json.dumps(data, indent = 4, cls =setEncoder)
    print(json_data)
```

```python
with open ('C:\\Users\\admin\\Desktop\\Assignment\\StudentsPerformance.csv','r') as f:
    reader = csv.reader(f)
    
    #skips header
    next(reader) 
    
    data = {"education levels":[],"scores":{"average":{}}}
    eList = []
    sList = []
    length = 0
    
    mScore =[]
    rScore = []
    wScore = []
    
    for row in reader:
        length+=1
        eList.append(row[2])
        mScore.append(row[5])
        rScore.append(row[6])
        wScore.append(row[7])
    
    #print(eList)   
    
    List =list(set(eList)) #convert set of unique education levels from SET to LIST
    #print(List)

    mScore = list(map(int,mScore))
    #print(mScore)
    rScore = list(map(int,rScore))
    #print(rScore)
    wScore = list(map(int,wScore))
    #print(wScore)
    
    mavg = sum(mScore)/length
    ravg = sum(rScore)/length
    wavg = sum(wScore)/length
    
    sList.append(mavg)
    sList.append(ravg)
    sList.append(wavg)
    #print(sList)
    
    
    for i in List:
        data["education levels"].append(i)
        
    data["scores"]["average"]["Math Score"]=mavg
    data["scores"]["average"]["Reading Score"]=ravg
    data["scores"]["average"]["Writing Score"]=wavg
    
    print(data)
```

```python
with open ("C:\\Users\\admin\\Desktop\\Assignment\\NewStudentsPerformance.json","w") as f:
    json_data = json.dumps(data, indent = 4, cls =setEncoder)
    print(json_data)
```
